# 1️⃣ Forward Propagation (Single Neuron)

```python
import numpy as np

# Given values
x = np.array([1, 2])        # inputs
w = np.array([0.4, 0.6])    # weights
b = 0.5                     # bias

# Linear combination
z = np.dot(w, x) + b

# Sigmoid activation
output = 1 / (1 + np.exp(-z))

print("z =", z)
print("Output =", output)
```

# 2️⃣ Backpropagation (Weight Update)

```python

learning_rate = 0.1
y_true = 1
y_pred = 0.8
x = 2
old_weight = 0.5

# Error
error = y_true - y_pred

# Weight update
delta_w = learning_rate * error * x
new_weight = old_weight + delta_w

print("Updated weight =", new_weight)
```

# 3️⃣ Mean Squared Error (MSE)

```python
import numpy as np

y_true = np.array([1, 0, 1])
y_pred = np.array([0.8, 0.2, 0.6])

mse = np.mean((y_true - y_pred) ** 2)

print("MSE =", mse)
```

# 4️⃣ Gradient Descent Numerical Example

```python
# Cost function: J(w) = (w - 3)^2

w = 5
learning_rate = 0.1

# Gradient
gradient = 2 * (w - 3)

# Update
w_new = w - learning_rate * gradient

print("Updated weight =", w_new)
```

# 5️⃣ Softmax Calculation

```python
import numpy as np

z = np.array([2, 1, 0])

# Softmax
softmax = np.exp(z) / np.sum(np.exp(z))

print("Softmax output =", softmax)
```

# 6️⃣ Number of Parameters in Neural Network

```python
# Network structure
input_neurons = 4
hidden_neurons = 3
output_neurons = 2

parameters = (input_neurons * hidden_neurons) + hidden_neurons + \
             (hidden_neurons * output_neurons) + output_neurons

print("Total parameters =", parameters)
```

# 7️⃣ One Training Step (Forward + Backward)

```python
# Simple neuron training step

x = 1
w = 0.5
b = 0.2
y_true = 1
lr = 0.1

# Forward
z = w * x + b
y_pred = 1 / (1 + np.exp(-z))

# Loss (MSE)
loss = (y_true - y_pred) ** 2

# Backward
dz = y_pred - y_true
dw = dz * x
db = dz

# Update
w -= lr * dw
b -= lr * db

print("Updated weight =", w)
print("Updated bias =", b)
```
